```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# PARAMETERS
OUTPUT_PATH = 'bt_data/'
NAMES = [0]

COLOR_A = 'C0'
COLOR_B = 'C9'
COLOR_C = 'C3'
```

```python
for name in NAMES:
    c = np.load(f'{OUTPUT_PATH}setup_{name}.npz')['c']
    plt.plot(c[:, 0], label=COLOR_A)
    plt.plot(c[:, 1], label=COLOR_B)
    plt.plot(c[:, 2], label=COLOR_C)
    plt.xlim(0, len(c))
    plt.ylim(0, 1)
    plt.legend()
    plt.show()
```
