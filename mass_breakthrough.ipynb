```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# PARAMETERS
OUTPUT_PATH = 'bt_data/'
NAMES = [0, 1]
VAR_LNK = [0.5, 0.5, 0.5, 1.0, 1.5]
CORR_LEN = [15, 20, 25, 25, 25]
TRANS_DISP = [0.5, 0.5, 0.5, 0.5, 0.5]

COLOR_A = 'C0'
COLOR_B = 'C9'
COLOR_C = 'C3'

INTERVAL = 8
N_SEED = 96
```

```python
for name in NAMES:
    c = np.load(f'{OUTPUT_PATH}setup_{name}.npz')['c']
    print(f'--------Plot setup {name} (N={N_SEED})--------')
    
    print(f'TMAX {COLOR_A}: {np.argmax(c[:,0])*INTERVAL}')
    print(f'CMAX {COLOR_A}: {np.max(c[:,0])}')
    print(f'STD {COLOR_A}: {np.std(c[:,0])}\n')
    plt.plot(c[:, 0], label='A', color=COLOR_A)
    plt.vlines(np.argmax(c[:, 0]), 0, 0.7, colors=COLOR_A)
    plt.hlines(np.max(c[:, 0]), 10+50, len(c), colors=COLOR_A)
    
    print(f'TMAX {COLOR_B}: {np.argmax(c[:,1])*INTERVAL}')
    print(f'CMAX {COLOR_B}: {np.max(c[:,1])}')
    print(f'STD {COLOR_B}: {np.std(c[:,1])}\n')
    plt.plot(c[:, 1], label='B', color=COLOR_B)
    plt.vlines(np.argmax(c[:, 1]), 0, 0.8, colors=COLOR_B)
    plt.hlines(np.max(c[:, 1]), 60+50, len(c), colors=COLOR_B)
    
    print(f'TMAX {COLOR_C}: {np.argmax(c[:,2])*INTERVAL}')
    print(f'CMAX {COLOR_C}: {np.max(c[:,2])}')
    print(f'STD {COLOR_C}: {np.std(c[:,2])}\n')
    plt.plot(c[:, 2], label='C', color=COLOR_C)
    plt.vlines(np.argmax(c[:, 2]), 0, 0.9, colors=COLOR_C)
    plt.hlines(np.max(c[:, 2]), 110+50, len(c), colors=COLOR_C)
    
    plt.text(np.argmax(c[:, 0])+3, 0.70, np.argmax(c[:, 0])*INTERVAL,rotation=0, color=COLOR_A)
    plt.text(np.argmax(c[:, 1])+3, 0.80, np.argmax(c[:, 1])*INTERVAL,rotation=0, color=COLOR_B)
    plt.text(np.argmax(c[:, 2])+3, 0.90, np.argmax(c[:, 2])*INTERVAL,rotation=0, color=COLOR_C)
    
    plt.text(10, np.max(c[:, 0]), round(np.max(c[:, 0]), 3),rotation=0, color=COLOR_A)
    plt.text(60, np.max(c[:, 1]), round(np.max(c[:, 1]), 3),rotation=0, color=COLOR_B)
    plt.text(110, np.max(c[:, 2]), round(np.max(c[:, 2]), 3),rotation=0, color=COLOR_C)
    
    plt.text(10, 0.95, r'$\alpha_T$'+f' = {TRANS_DISP[name]}m')
    plt.text(10, 0.9, r'$L(m)$'+f' = {CORR_LEN[name]}m')
    plt.text(10, 0.85, r'$\sigma_k$'+f' = {VAR_LNK[name]}')
    
    plt.title(f'Breakthrough time for setup {name} (N={N_SEED})')
    plt.xlim(0, len(c))
    plt.ylim(0, 1)
    plt.xlabel('Saved time step')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.show()
```
