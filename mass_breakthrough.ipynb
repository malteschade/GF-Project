```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# PARAMETERS
OUTPUT_PATH = 'bt_data/'
NAMES = [0, 1]

COLOR_A = 'C0'
COLOR_B = 'C9'
COLOR_C = 'C3'

INTERVAL = 8
```

```python
for name in NAMES:
    c = np.load(f'{OUTPUT_PATH}setup_{name}.npz')['c']
    print(f'--------Plot setup {name}--------')
    
    print(f'TMAX {COLOR_A}: {np.argmax(c[:,0])*INTERVAL}')
    print(f'CMAX {COLOR_A}: {np.max(c[:,0])}')
    print(f'STD {COLOR_A}: {np.std(c[:,0])}\n')
    plt.plot(c[:, 0], label=COLOR_A, color=COLOR_A)
    plt.vlines(np.argmax(c[:, 0]), 0, 1, colors=COLOR_A)
    
    print(f'TMAX {COLOR_B}: {np.argmax(c[:,1])*INTERVAL}')
    print(f'CMAX {COLOR_B}: {np.max(c[:,1])}')
    print(f'STD {COLOR_B}: {np.std(c[:,1])}\n')
    plt.plot(c[:, 1], label=COLOR_B, color=COLOR_B)
    plt.vlines(np.argmax(c[:, 1]), 0, 1, colors=COLOR_B)
    
    print(f'TMAX {COLOR_C}: {np.argmax(c[:,2])*INTERVAL}')
    print(f'CMAX {COLOR_C}: {np.max(c[:,2])}')
    print(f'STD {COLOR_C}: {np.std(c[:,2])}\n')
    plt.plot(c[:, 2], label=COLOR_C, color=COLOR_C)
    plt.vlines(np.argmax(c[:, 2]), 0, 1, colors=COLOR_C)
    
    plt.xlim(0, len(c))
    plt.ylim(0, 1)
    plt.legend()
    plt.show()
```
